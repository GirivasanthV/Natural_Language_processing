```python
import re
import nltk
from nltk.corpus import udhr, stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# ✅ Set NLTK data path (VERY IMPORTANT)
nltk.data.path.append('/Users/girivasanth/nltk_data')

# Tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Load UD English text
text = udhr.raw('English-Latin1')[:5000]

# Normalization
text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)

# ✅ Tokenization (offline-safe)
tokens = text.split()

# Stopword Removal
tokens = [t for t in tokens if t not in stop_words]

# Stemming
stemmed = [stemmer.stem(t) for t in tokens]

# Lemmatization
lemmatized = [lemmatizer.lemmatize(t) for t in tokens]

# Output
print("UD Tokens:", tokens[:20])
print("UD Stemmed:", stemmed[:20])
print("UD Lemmatized:", lemmatized[:20])
```
