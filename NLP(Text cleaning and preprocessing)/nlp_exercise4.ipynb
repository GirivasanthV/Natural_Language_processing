```python
import re
import nltk
from nltk.corpus import reuters, stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# ✅ Set NLTK data path (IMPORTANT)
nltk.data.path.append('/Users/girivasanth/nltk_data')

# Tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Load Reuters corpus
text = " ".join(reuters.words()[:2000])

# Normalization
text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)

# ✅ Tokenization (offline-safe)
tokens = text.split()

# Stopword Removal
tokens = [t for t in tokens if t not in stop_words]

# Stemming
stemmed = [stemmer.stem(t) for t in tokens]

# Lemmatization
lemmatized = [lemmatizer.lemmatize(t) for t in tokens]

# Output
print("Reuters Tokens:", tokens[:20])
print("Reuters Stemmed:", stemmed[:20])
print("Reuters Lemmatized:", lemmatized[:20])
```
