```python
import re
import nltk
from sklearn.datasets import fetch_20newsgroups
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# ✅ Explicit NLTK data path
nltk.data.path.append('/Users/girivasanth/nltk_data')

# Tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Load dataset
data = fetch_20newsgroups(subset='train')
text = data.data[0]

# Normalization
text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)

# ✅ Tokenization (NO punkt / punkt_tab)
tokens = text.split()

# Stopword Removal
tokens = [t for t in tokens if t not in stop_words]

# Stemming
stemmed = [stemmer.stem(t) for t in tokens]

# Lemmatization
lemmatized = [lemmatizer.lemmatize(t) for t in tokens]

# Output
print("20 News Tokens:", tokens[:20])
print("20 News Stemmed:", stemmed[:20])
print("20 News Lemmatized:", lemmatized[:20])
```
